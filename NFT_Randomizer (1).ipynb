```python
#NOTE: if you want differente basic models and not just 1, make the Background default model image
# a void transparent one (an empty .png) and create a folder for the models, so you will be choosing
# one of the models in a random way like the rest of the elements for the final image and you will 
# get the desired result


#Importing Image class from PIL package
from PIL import Image
import os, os.path

#Import random module
import random


#Background image (called 1.png) will be always the same (default model)
background = Image.open(r"C:\NFTImages\1.png")
#background.show()

#Explore all the folders in the root folder
path = "C:/NFTImages/"
print("Elements",os.listdir(path))
print("Total Elements",len(os.listdir(path)))

#This loop will iterate over every folder in the root folder, 
#selecting one image in each one and adding it to the final image
for item in os.listdir(path):
    
    #Result folder must be ignored
    if item!="Result":
    
        #Creating the path of the folder
        destinationPath=(os.path.join(path, item))

        #If the item is a folder then we will take one of the contained images
        if os.path.isdir(destinationPath): 
            print("IT IS A folder")
            print("Elements",os.listdir(destinationPath))

            #Random number (between 1 and max elements in folder)
            #to get the image that will be used
            r1 = random.randint(0, len(os.listdir(destinationPath))-1)
            print("random number chosen",r1)
            selectedImg = os.listdir(destinationPath)[r1]
            print(selectedImg)

            #adds / to path
            imagePath1=(os.path.join(destinationPath, ''))
            print(imagePath1)
            print(destinationPath)

            imagePath2=os.path.join(imagePath1, selectedImg)
            print(imagePath2)

            foreground = Image.open(imagePath2)
            background.paste(foreground, (0, 0), foreground)
            #background.show()


        #Load all the images in the folder to make a random and take only one
        #Empty list to append the final image step by step
        imgs = []
        filename = None
        valid_images = [".jpg",".gif",".png",".tga"]
        for f in os.listdir(path):
            ext = os.path.splitext(f)[1]
            if ext.lower() not in valid_images:
                continue
            imgs.append(Image.open(os.path.join(path,f)))

        print(imgs)

        
#Path where the final Image will be stored
destinationPath = "C:/NFTImages/Result/"
    
#Creating name
ext=str(len(os.listdir(destinationPath)))+".png"
        
#Image saved in the destination path with the name and extension previously created
background.save(os.path.join(destinationPath, ext))
```
